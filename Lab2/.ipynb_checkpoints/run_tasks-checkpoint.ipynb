```python
import pandas as pd

from matplotlib.pyplot import *
from matplotlib import pylab
from numpy import * 
from time import clock
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.externals import joblib
from sklearn.model_selection import train_test_split

from utils import *
from binary_relevance import BR
from label_powerset import LP

set_printoptions(precision=2)
```

```python
########################
# DATA PREPARATION
########################

print("Load Data")

#df = pd.read_csv("music.csv")
#L = 6

df = pd.read_csv("imdb_vectorized.csv.gz")
L = 28

labels = array(df.columns.values.tolist())[0:L]
data = df.values

print("Split Data")
N = int(len(data)*2/3)

Y = data[:,0:L].astype(float)
X = data[:,L:].astype(float)
```

```python
########################
# EXPLORATION
########################

print("Task 1. Exploration")

label_cardinality = np.sum(Y, axis=1)
label_density = divide(label_cardinality,ones(Y.shape[0])*L-label_cardinality)
label_relevance_frequency = np.mean(Y, axis=1)
label_freqs = np.mean(Y, axis=0)

label_unique_with_count = count_combinations(Y)
label_unique = np.array([np.fromstring(label, sep=' ') for label in list(label_unique_with_count.keys())])
combinations_freq = label_unique_with_count.values()
unique_combinations = label_unique.shape[0]

#HINT (plot 1): 
figure()
plot(range(L),-sort(-label_freqs))
xticks(range(L), labels[argsort(-label_freqs)], rotation=45)
xlabel("Labels")
ylabel("Frequency")
show()

# HINT (plot 2): 
figure()
xlabel("Label combinations")
ylabel("Frequency")
# (log scale on y axis)
semilogy(range(unique_combinations), combinations_freq)
#savefig("frec_"+str(L)+".pdf")
show()
```

```python
print("Task 2. Marginal Dependence")

def marginal_dependence(M, min_weight=0.1):
    top = min(10,L) # Deal with the 10 most frequent labels
    idx = argsort(-sum(M,axis=0))[:top]
    S = corrcoef(M[:,idx], rowvar=False)
    fig = make_heatmap(S,labels[idx])
    show()

    G = make_graph(S,labels[idx], min_weight=min_weight)
    
marginal_dependence(Y)
```

```python
print("Task 3. Conditional Dependence")

print("    Logistic regression")
lr_model = BR(LogisticRegression(solver='liblinear'))

lr_E = get_errors(lr_model,X,Y,2)
marginal_dependence(lr_E, min_weight=0.03)
```

```python
print("    Decision Tree Classifier")
tree_model = BR(DecisionTreeClassifier())

tree_E = get_errors(tree_model,X,Y,2)
marginal_dependence(tree_E, min_weight=0.03)
```

```python
########################
# CLASSIFICATION
########################

print("Task 4. Evaluate LP vs BR (hamming loss, 0/1 loss, running time.")

def run_model(model):
    temp = clock()
    E = get_errors(model,X,Y,2)
    zero_one_loss = sum(1-E)/E.size
    hamming_loss = sum([not row.all() for row in E])/E.shape[0]

    return clock()-temp, zero_one_loss, hamming_loss

BR_perf = run_model(BR(LogisticRegression(solver='liblinear')))
LP_perf = run_model(LP(LogisticRegression(solver='liblinear')))
```

```python
log = figure()

data = [(a,b) for a,b in zip(BR_perf, LP_perf)]

legends = ["BR", "LR"]
objects = ("Time", "0/1", "Hamming")
y_pos = np.arange(len(objects))
pylab.title("LR BR comparision")
pylab.gca().set_yscale('log')

dim = len(data[0])
w = 0.75
dimw = w / dim

x = pylab.arange(len(data))
for i in range(len(data[0])) :
    y = [d[i] for d in data]
    b = pylab.bar(x + i * dimw, y, dimw, bottom=0.001, label=legends[i])
xticks(y_pos, objects)
legend()
show()
```
